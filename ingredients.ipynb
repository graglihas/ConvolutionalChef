```
import numpy as np
import pandas as pd
import csv
import regex as re
title = []
total = []
```

```
with open("ingredients200.txt","r",encoding='cp1252') as filename:
  for line in filename:
    currentline = line.split(",")
    total += currentline[1:]
    title += [currentline[0]]
```

```
ingr = []
temp=[]
temp1=[]
temp2=[]
temp3=[]
temp4=[]
for i in total:
  temp.append(i)
  g=i
  j=g.rfind('-')
  if j!=-1:
    g=g[:j-1]
  temp2.append(g)
  g='0 '+g
  k=len(g)-re.search('\d',g[::-1]).start(0)
  g=g[k+1:]
  #print(g)
  temp3.append(g)
  g=g.lower()
  temp4.append(g)
  g=g.replace('few ','').replace('cups ','').replace('cup ','').replace('tablespoons ','').replace('tablespoon ','').replace('grams ','').replace('gram ','').replace('teaspoons ','').replace('teaspoon ','').replace('tbsp ','').replace('lb ','').replace('stalks ','').replace('stalk ','').replace('fresh ','').replace('veg ','').replace('as much as ','').replace('sprigs ','').replace('sprig ','').replace('pinch ','').replace('inch ','').replace('sprig of ','').replace('whole ','').replace('ml ','').replace('raw ','').replace('pinch ','').replace('tsp ','').replace('kg ','').replace('cloves ','').replace('clove ','').replace('inches ','').replace('big ','').replace('spoon ','')
  g=g.strip()
  ingr.append(g)
ing=(list(set(ingr)))
df1 = pd.DataFrame(ing)
df = pd.DataFrame(list(zip(temp,temp2,temp3,temp4,ingr)))
df.to_csv('ing200.csv',index=False)
df1.to_csv('unique_ing200.csv',index=False)
```
